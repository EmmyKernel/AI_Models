#### Question 2

```python
def is_even(k):
    k = abs(k)
    while k > 1:
        k = k - 2
    return k == 0


print(is_even(4))
print(is_even(7))   
print(is_even(0))   
print(is_even(-6)) 
```

#### Question 1

```python
def is_multiple(n, m):
    if m == 0:
        return "Cannot check multiples of zero"
    return n % m == 0

print(is_multiple(10, 5))
print(is_multiple(15, 2))
print(is_multiple(0, 3))
print(is_multiple(-12, 0))
```

#### Question 3

```python
while True:
    print("\nEnter 5 numbers:")
    numbers = []

    for i in range(5):
        try:
            num = float(input(f"Enter number {i+1}: "))
            numbers.append(num)
        except ValueError:
            print("Invalid input! Please enter a valid number.")
            break
    else:
        average = sum(numbers) / len(numbers)
        print(f"The average of {numbers} is: {average:.2f}")
    
    continue_choice = input("\nDo you want to enter another set of 5 numbers? (yes/n): ")
    if continue_choice.lower() != 'yes':
        print("Goodbye!")
        break
```
