#### Question 6

```python
sample_list = [2,4,5,8,10]

def multiple_n():
    multiplied_list = []
    for n in sample_list:
        multiplied_value = n * 3
        multiplied_list.append(multiplied_value)
    return multiplied_list

new_list = multiple_n()
new_list
```

#### Question 2

```python
def is_even(k):
    k = abs(k)
    while k > 1:
        k = k - 2
    return k == 0


print(is_even(4))
print(is_even(7))   
print(is_even(0))   
print(is_even(-6)) 
```
